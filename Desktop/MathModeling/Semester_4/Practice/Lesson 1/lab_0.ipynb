```python
def prime(x: int) -> bool:
    """ Проверка числа на простоту

    :param x: Вводимое положительное число
    :type x: int
    :return: True - если число простое, False - если не простое
    :rtype: bool
    """
    if x < 2:
        return False
    for i in range(2, round(x**0.5)+1):
        if x % i == 0:
            return False
    return True

counter = 0
print("Простые числа: ", end="")
for i in range(1, 1001):
    if prime(i):
        print(i, end=", ")
        counter += 1

print(f"\nКоличество простых чисел: {counter}")
```
