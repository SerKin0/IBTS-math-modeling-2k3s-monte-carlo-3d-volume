## Задание 1

**Вводные данные**

```python
import numpy as np

X = np.array([
    [1, 0, 1],
    [2, 0, 2],
    [3, 0, 3],
    [4, 4, 4]
])

result = np.prod([i for i in X.diagonal() if i != 0])
print(f"{result=}")
```
